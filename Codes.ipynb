```python
def cleanReview(text):
    text=re.sub(r'@[A-Za-z0-9]+', '', text)  #removing @mention
    text=re.sub(r'#', '', text) #Removing the '#' symbols
    text=re.sub(r'RT[\s]+', '', text)#Removing RT
    text=re.sub(r'https?:\/\/\S+', '',text) #Removing hyper links
    
    return text
library_df['content'] = library_df['content'].apply(cleanReview)
```
